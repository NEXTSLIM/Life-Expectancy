```python
import pandas as pd
import numpy as np
```

```python
ml= "ML.csv"
ml_df = pd.read_csv(ml)
ml_df.head()
```

```python
ml_df.dtypes
```

```python
ml_df=ml_df.drop(columns=['IndicatorCode'])
ml_df.head()
```

```python
indicators = ml_df.pivot_table('Value', ['CountryName','CountryCode','Year', 'Region'], 'IndicatorName')
indicators
```

```python

indicators_df = indicators.rename(columns={"CountryName":"Country","CountryCode":"Code"})

indicators_df
```

```python
indicators.to_csv ('cleaned_ML_data_2.csv', index = True, header=True)
```
