```python
import pandas as pd
import numpy as np
```

```python
gender= "cleaned_gender_2019.csv"
gender_df = pd.read_csv(gender)
gender_df.head()
```

```python
ml= "cleaned_ML_data_3.csv"
ml_df = pd.read_csv(ml)
ml_df.head()
```

```python
s1 = pd.merge(ml_df,gender_df, how='left', on=['Country','Code','Year'])
s1
```

```python
s1.to_csv ('combined_ml_genders.csv', index = False, header=True)
```
